```python
#I take as date 12/02/2022 because is seems that there was not any commits in the last two weeks

params_dict= {'auth': ('AlbaNMG',key),'since' : '?2022-02-12T00:00:00Z'}

url='https://api.github.com/repos/ta-data-lis/lab-mysql-first-queries/commits'
```

```python
#request=requests.get(url, auth=('AlbaNMG',key))
request=requests.get(url, params=params_dict)
```

```python
print("request:", request.status_code)
```

```python
data2=request.json()
data2
```

```python
#chacking the columns

dataframe2=pd.DataFrame(data2)
dataframe2.columns
```

```python
dataframe2[['author', 'commit','committer']]
```

```python
challenge2=pd.json_normalize(dataframe2['commit'])
challenge2
```

```python
#counting rows
x=len(challenge2.index)
x

#There were 3 commits in that lab since 12/02/2022
```
